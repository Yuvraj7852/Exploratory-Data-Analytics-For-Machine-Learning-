Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('old car price priduction.csv')
```

```python
dataset.head()
```

Exploratory Data Asnalysis


```python
#Check Missing value in dataset
dataset.isnull().sum()
```

```python
#delete id-column in dataset
dataset=dataset.drop(columns='id')
```

```python
dataset.head(5)
```

```python
#standardizing Nomeric Formats-:
#Currency Symbols-:
dataset['car_prices_in_rupee'] = dataset['car_prices_in_rupee'].str.replace(" Lakh","000")
dataset['car_prices_in_rupee'] = dataset['car_prices_in_rupee'].str.replace(" Crore","0000")
dataset['car_prices_in_rupee'] = dataset['car_prices_in_rupee'].str.replace(".","")
dataset['car_prices_in_rupee'] = dataset['car_prices_in_rupee'].str.replace(",","")

#Kms Symbols-:
dataset['kms_driven'] = dataset['kms_driven'].str.replace("kms","")
dataset['kms_driven'] = dataset['kms_driven'].str.replace(",","")

#normal form
dataset['ownership'] = dataset['ownership'].str.replace("st Owner","")
dataset['ownership'] = dataset['ownership'].str.replace("nd Owner","")
dataset['ownership'] = dataset['ownership'].str.replace("rd Owner","")
dataset['ownership'] = dataset['ownership'].str.replace("th Owner","")


dataset['engine'] = dataset['engine'].str.replace(" cc","")

dataset['Seats'] = dataset['Seats'].str.replace(" Seats","")
```

```python
dataset.head(5)
```

```python
dataset.info()
```

```python
#converd data type  in int\float-:
dataset['car_prices_in_rupee']=dataset['car_prices_in_rupee'].astype("int64")
dataset['kms_driven']=dataset['kms_driven'].astype("int64")
dataset['ownership']=dataset['ownership'].astype("int64")
dataset['engine']=dataset['engine'].astype("int64")
dataset['Seats']=dataset['Seats'].astype("int64")
dataset.info()
```

```python
#String data covert in int by LabelEncoder
from sklearn.preprocessing import LabelEncoder 

ver=LabelEncoder()
dataset["fuel_type"]=ver.fit_transform(dataset["fuel_type"])

dataset["transmission"]=ver.fit_transform(dataset["transmission"])

dataset["car_name"]=ver.fit_transform(dataset["car_name"])
```

```python
dataset.info()
```
